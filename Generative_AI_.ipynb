Question 3: What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?

Answer:Autoencoders (AE):

Autoencoders are neural networks that learn to compress (encode) input data into a lower-dimensional representation and then reconstruct (decode) it back to the original form.

- Work deterministically (fixed encoding for each input)
- Learn a single latent representation (vector)
- Mainly used for feature extraction and reconstruction
- Not ideal for generating diverse new text

In text: An AE can reconstruct a sentence but struggles to generate new, varied sentences

Variational Autoencoders (VAE):

Variational Autoencoders are a probabilistic version of autoencoders that learn a distribution over the latent space instead of a single point.

- Learn mean (μ) and variance (σ) → latent space becomes continuous
- Use sampling to generate new data
- Introduce randomness → better for generating new text
- Ensure smooth and meaningful latent space

 In text: A VAE can generate different sentences by sampling from the latent distribution

Key Differences:

Feature	Autoencoder (AE)	Variational Autoencoder (VAE)
Nature	Deterministic	Probabilistic
Latent Space	Fixed vector	Distribution (μ, σ)
Output	Reconstruction only	Reconstruction + generation
Diversity	Low	High
Text Generation	Weak	Strong

Simple Explanation:

AE: “Input ko compress karke wapas same banana”
VAE: “Data ka pattern samajh ke us jaisa naya data banana”

Summary:

Autoencoders learn deterministic latent representations and are mainly used for reconstruction, while Variational Autoencoders learn probabilistic latent distributions, enabling them to generate diverse and new text samples.

Question 4: Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

Answer: Working of Attention Mechanism in NMT:

In Neural Machine Translation (NMT), the attention mechanism helps the model focus on relevant parts of the input sentence while generating each word of the output sentence.

Step-by-step working:

1) Encoder:

- Processes the input sentence and converts each word into hidden states
- Example: “I love India” → multiple hidden vectors

Decoder:

-Generates the translated sentence word by word

3)Attention Mechanism:

- At each decoding step, the model:
  - Compares the current decoder state with all encoder hidden states
  - Assigns attention weights (importance scores) to each input word
  - Creates a context vector (weighted sum of encoder outputs)

4)Final Output:

- The decoder uses this context vector to generate the next word

 Example:

While translating “I love India” → “मुझे भारत पसंद है”
When generating “भारत”, the model focuses more on “India”

Why Attention is Critical:

1) Handles Long Sentences

 - Without attention, models forget earlier words
 - Attention allows access to the entire input sequence

2) Improves Translation Quality

- Focuses on the most relevant words
- Produces more accurate and meaningful translations

3) Solves Bottleneck Problem

- Traditional models use a single fixed context vector
- Attention provides dynamic context at every step

4)Better Word Alignment

- Learns which input words correspond to output words

Simple Explanation:

Attention = “Model ka focus system”

 Har word generate karte time model decide karta hai ki input ke kis part par dhyan dena hai

Summary:

Attention mechanisms in NMT allow the decoder to focus on relevant parts of the input sequence by assigning weights to encoder outputs at each step, improving translation accuracy, handling long sentences, and overcoming the limitations of fixed context vectors.

Answer : Below is a simple example of a Variational Autoencoder (VAE) for text reconstruction using the given small dataset.

Explanation

A VAE has 3 main parts:

Encoder

It takes the padded sentence sequence and converts it into a latent
representation using:

z_mean

z_log_var

Sampling Layer

It samples a latent vector z from the learned distribution.
Decoder
It reconstructs the sentence from the latent vector.

Since the dataset is very small, this model is only for demonstration. It may reconstruct close sentences, but not perfect language generation like large NLP models.

Python Code
# Simple Variational Autoencoder (VAE) for text reconstruction
# Dataset:
# ["The sky is blue", "The sun is bright", "The grass is green",
#  "The night is dark", "The stars are shining"]

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, Model

# -------------------------------
# 1. Dataset
# -------------------------------
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# -------------------------------
# 2. Preprocessing
# -------------------------------
tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)
max_len = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

word_index = tokenizer.word_index
index_word = {v: k for k, v in word_index.items()}

vocab_size = len(word_index) + 1  # +1 for padding token

print("Word Index:")
print(word_index)
print("\nTokenized Sequences:")
print(sequences)
print("\nPadded Sequences:")
print(padded_sequences)

# -------------------------------
# 3. Prepare input/output
# -------------------------------
x_train = padded_sequences
y_train = np.expand_dims(padded_sequences, -1)  # shape: (samples, seq_len, 1)

# -------------------------------
# 4. Sampling Layer
# -------------------------------
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        epsilon = tf.random.normal(shape=tf.shape(z_mean))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# -------------------------------
# 5. Encoder
# -------------------------------
embedding_dim = 16
latent_dim = 8

encoder_inputs = layers.Input(shape=(max_len,), name="encoder_input")
x = layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim)(encoder_inputs)
x = layers.Flatten()(x)
x = layers.Dense(32, activation="relu")(x)

z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])

encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# -------------------------------
# 6. Decoder
# -------------------------------
latent_inputs = layers.Input(shape=(latent_dim,), name="z_sampling")
x = layers.Dense(32, activation="relu")(latent_inputs)
x = layers.Dense(max_len * vocab_size, activation="linear")(x)
decoder_outputs = layers.Reshape((max_len, vocab_size))(x)

decoder = Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# -------------------------------
# 7. VAE Model
# -------------------------------
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.loss_tracker = tf.keras.metrics.Mean(name="loss")
        self.reconstruction_loss_tracker = tf.keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [self.loss_tracker, self.reconstruction_loss_tracker, self.kl_loss_tracker]

    def train_step(self, data):
        x, y = data

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(x)
            reconstruction_logits = self.decoder(z)

            # Reconstruction loss
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    tf.keras.losses.sparse_categorical_crossentropy(
                        y, reconstruction_logits, from_logits=True
                    ),
                    axis=1
                )
            )

            # KL divergence loss
            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var), axis=1)
            )

            total_loss = reconstruction_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        return self.decoder(z)

vae = VAE(encoder, decoder)
vae.compile(optimizer=tf.keras.optimizers.Adam())

# -------------------------------
# 8. Train the model
# -------------------------------
vae.fit(x_train, y_train, epochs=300, batch_size=2, verbose=0)

print("\nTraining complete!")

# -------------------------------
# 9. Helper function to decode sequence
# -------------------------------
def sequence_to_text(sequence):
    words = []
    for idx in sequence:
        if idx != 0 and idx in index_word:
            words.append(index_word[idx])
    return " ".join(words)

# -------------------------------
# 10. Reconstruction
# -------------------------------
print("\n--- Reconstruction Results ---")
reconstructed_logits = vae.predict(x_train, verbose=0)
reconstructed_sequences = np.argmax(reconstructed_logits, axis=-1)

for original, recon_seq in zip(sentences, reconstructed_sequences):
    reconstructed_text = sequence_to_text(recon_seq)
    print(f"Original:      {original}")
    print(f"Reconstructed: {reconstructed_text}")
    print()

# -------------------------------
# 11. Generate similar sentences
# -------------------------------
print("\n--- Generated Sentences from Random Latent Vectors ---")
for i in range(3):
    random_z = np.random.normal(size=(1, latent_dim))
    generated_logits = decoder.predict(random_z, verbose=0)
    generated_sequence = np.argmax(generated_logits, axis=-1)[0]
    generated_text = sequence_to_text(generated_sequence)
    print(f"Generated {i+1}: {generated_text}")

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.

Answer : Below is a simple attention-based encoder-decoder model for English-to-Spanish translation using TensorFlow/Keras.
This is a toy example with a very small dataset, so it is only for understanding the working of the model.

Explanation

This model has 3 main parts:

1. Encoder
Takes the English input sentence
Converts words into embeddings
Passes them through an LSTM
Produces:
encoder outputs
hidden state
cell state
2. Attention
At each decoding step, attention looks at all encoder outputs
It decides which input words are more important
It creates a context vector
3. Decoder
Takes:
previous Spanish word
context vector from attention
previous hidden state
Predicts the next Spanish word
Python Code
# Simple Attention-based Encoder-Decoder for English-to-Spanish Translation
# TensorFlow / Keras implementation

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Layer, Embedding, LSTM, Dense
from tensorflow.keras import Model

# --------------------------------------------------
# 1. Small toy dataset
# --------------------------------------------------
english_sentences = [
    "hello",
    "how are you",
    "i am fine",
    "good morning",
    "thank you",
    "i love you",
    "see you soon",
    "what is your name"
]

spanish_sentences = [
    "inicio hola fin",
    "inicio como estas fin",
    "inicio estoy bien fin",
    "inicio buenos dias fin",
    "inicio gracias fin",
    "inicio te amo fin",
    "inicio hasta pronto fin",
    "inicio como te llamas fin"
]

# --------------------------------------------------
# 2. Tokenization
# --------------------------------------------------
eng_tokenizer = Tokenizer(filters='')
eng_tokenizer.fit_on_texts(english_sentences)
eng_sequences = eng_tokenizer.texts_to_sequences(english_sentences)
eng_vocab_size = len(eng_tokenizer.word_index) + 1

spa_tokenizer = Tokenizer(filters='')
spa_tokenizer.fit_on_texts(spanish_sentences)
spa_sequences = spa_tokenizer.texts_to_sequences(spanish_sentences)
spa_vocab_size = len(spa_tokenizer.word_index) + 1

max_eng_len = max(len(seq) for seq in eng_sequences)
max_spa_len = max(len(seq) for seq in spa_sequences)

encoder_input_data = pad_sequences(eng_sequences, maxlen=max_eng_len, padding='post')
decoder_input_data = pad_sequences([seq[:-1] for seq in spa_sequences], maxlen=max_spa_len-1, padding='post')
decoder_target_data = pad_sequences([seq[1:] for seq in spa_sequences], maxlen=max_spa_len-1, padding='post')

# reshape target for sparse_categorical_crossentropy
decoder_target_data = np.expand_dims(decoder_target_data, -1)

print("Encoder input shape:", encoder_input_data.shape)
print("Decoder input shape:", decoder_input_data.shape)
print("Decoder target shape:", decoder_target_data.shape)

# --------------------------------------------------
# 3. Attention Layer
# --------------------------------------------------
class AttentionLayer(Layer):
    def __init__(self):
        super(AttentionLayer, self).__init__()

    def call(self, decoder_hidden, encoder_outputs):
        # decoder_hidden shape: (batch, hidden)
        # encoder_outputs shape: (batch, seq_len, hidden)

        # Expand decoder hidden for score calculation
        decoder_hidden_time = tf.expand_dims(decoder_hidden, 1)

        # Score = dot product
        score = tf.reduce_sum(encoder_outputs * decoder_hidden_time, axis=2)

        # Attention weights
        attention_weights = tf.nn.softmax(score, axis=1)

        # Context vector
        attention_weights_time = tf.expand_dims(attention_weights, 2)
        context_vector = attention_weights_time * encoder_outputs
        context_vector = tf.reduce_sum(context_vector, axis=1)

        return context_vector, attention_weights

# --------------------------------------------------
# 4. Encoder
# --------------------------------------------------
class Encoder(Model):
    def __init__(self, vocab_size, embedding_dim, enc_units):
        super(Encoder, self).__init__()
        self.enc_units = enc_units
        self.embedding = Embedding(vocab_size, embedding_dim)
        self.lstm = LSTM(enc_units, return_sequences=True, return_state=True)

    def call(self, x):
        x = self.embedding(x)
        output, state_h, state_c = self.lstm(x)
        return output, state_h, state_c

# --------------------------------------------------
# 5. Decoder
# --------------------------------------------------
class Decoder(Model):
    def __init__(self, vocab_size, embedding_dim, dec_units):
        super(Decoder, self).__init__()
        self.dec_units = dec_units
        self.embedding = Embedding(vocab_size, embedding_dim)
        self.lstm = LSTM(dec_units, return_sequences=True, return_state=True)
        self.fc = Dense(vocab_size)
        self.attention = AttentionLayer()

    def call(self, x, hidden_h, hidden_c, encoder_outputs):
        # x shape: (batch, 1)
        x = self.embedding(x)  # (batch, 1, embedding_dim)

        # Attention
        context_vector, attention_weights = self.attention(hidden_h, encoder_outputs)
        context_vector = tf.expand_dims(context_vector, 1)

        # Concatenate context + embedding
        x = tf.concat([context_vector, x], axis=-1)

        output, state_h, state_c = self.lstm(x, initial_state=[hidden_h, hidden_c])
        output = tf.reshape(output, (-1, output.shape[2]))
        output = self.fc(output)

        return output, state_h, state_c, attention_weights

# --------------------------------------------------
# 6. Hyperparameters
# --------------------------------------------------
embedding_dim = 64
units = 128
batch_size = len(english_sentences)
epochs = 300

encoder = Encoder(eng_vocab_size, embedding_dim, units)
decoder = Decoder(spa_vocab_size, embedding_dim, units)

optimizer = tf.keras.optimizers.Adam()
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')

def loss_function(real, pred):
    loss_ = loss_object(real, pred)
    mask = tf.cast(tf.not_equal(real, 0), dtype=loss_.dtype)
    loss_ *= mask
    return tf.reduce_mean(loss_)

# --------------------------------------------------
# 7. Training step
# --------------------------------------------------
@tf.function
def train_step(enc_inp, dec_inp, dec_tar):
    loss = 0

    with tf.GradientTape() as tape:
        encoder_outputs, enc_h, enc_c = encoder(enc_inp)

        dec_h, dec_c = enc_h, enc_c

        for t in range(dec_inp.shape[1]):
            current_input = tf.expand_dims(dec_inp[:, t], 1)
            predictions, dec_h, dec_c, _ = decoder(current_input, dec_h, dec_c, encoder_outputs)

            loss += loss_function(dec_tar[:, t], predictions)

    batch_loss = loss / int(dec_inp.shape[1])

    variables = encoder.trainable_variables + decoder.trainable_variables
    gradients = tape.gradient(loss, variables)
    optimizer.apply_gradients(zip(gradients, variables))

    return batch_loss

# --------------------------------------------------
# 8. Train model
# --------------------------------------------------
for epoch in range(epochs):
    batch_loss = train_step(
        tf.constant(encoder_input_data),
        tf.constant(decoder_input_data),
        tf.constant(decoder_target_data[:, :, 0])
    )

    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch+1}, Loss: {batch_loss.numpy():.4f}")

# --------------------------------------------------
spa_index_word = {v: k for k, v in spa_tokenizer.word_index.items()}
eng_word_index = eng_tokenizer.word_index

# 10. Translation function

def translate_sentence(sentence):

    seq = pad_sequences(seq, maxlen=max_eng_len, padding='post')

    encoder_outputs, enc_h, enc_c = encoder(tf.constant(seq))

    dec_h, dec_c = enc_h, enc_c
    start_token = spa_tokenizer.word_index['inicio']
    end_token = spa_tokenizer.word_index['fin']

    dec_input = tf.constant([[start_token]])
    result = []

    for _ in range(max_spa_len):
        predictions, dec_h, dec_c, attention_weights = decoder(dec_input, dec_h, dec_c, encoder_outputs)
        predicted_id = tf.argmax(predictions[0]).numpy()

        if predicted_id == end_token or predicted_id == 0:
            break

        result.append(spa_index_word.get(predicted_id, ''))
        dec_input = tf.constant([[predicted_id]])

    return " ".join(result)

Question 10: Imagine you are building a creative writing assistant for a publishing company. The assistant should generate story plots and character descriptions using Generative AI. Answer: To build a creative writing assistant for a publishing company, I would design a system that can generate story ideas, plot outlines, character profiles, and writing suggestions while keeping the output creative, safe, and useful for human writers.

1. System Goal

The main purpose of the system would be to assist writers by:

- generating story plots
- creating character descriptions
- suggesting settings, conflicts, and endings
- helping with genre-specific writing such as mystery, romance, fantasy, or science fiction

This system should act as a support tool for writers, not a complete replacement for human creativity.

2. Model Selection

For this task, I would use a pre-trained Large Language Model (LLM) such as GPT-style or Transformer-based generative models.

Why this type of model?

Because story writing requires:

- understanding of language
- long-form text generation
- context retention
- creativity and style adaptation

Possible model choice

- GPT-type model for text generation
- Fine-tuned transformer model for publishing-specific needs
- Optionally, a retrieval-augmented system to bring in genre rules, writing - - - guidelines, or company style preferences

System design idea

The system can have multiple modules:

1.Plot Generator – creates story ideas and outlines

2.Character Generator – builds names, personalities, goals, and backgrounds

3.Style Controller – adjusts tone, genre, age group, and writing style

4.Safety Filter – checks harmful, biased, or inappropriate content

3. Training Data

The quality of the assistant depends heavily on the training data.

Types of data needed

- novels and short stories from different genres

- plot summaries

- character sketches

- screenplay outlines

- editorial writing examples

publishing house guidelines and style documents

Important points

The dataset should be:

- diverse in genre and writing style
- legally licensed or copyright-safe
- balanced across cultures and perspectives
- cleaned to remove low-quality or harmful text

Example data format

Input:

“Generate a mystery plot set in Delhi with a female detective”

Output:

a story summary with conflict, twists, and ending idea

This kind of prompt-response dataset is useful for fine-tuning.

4. Bias Mitigation

Bias is a major issue in generative AI because the model learns from human-written data, which may contain stereotypes or unfair patterns.

Steps to reduce bias

1.Careful dataset curation

- remove offensive or highly biased examples

- include diverse characters, cultures, and voices

2.Bias testing

- test prompts related to gender, caste, race, religion, age, disability, etc.

- check whether the model produces stereotypes

3.Safety rules and moderation

-add filters to block hateful, discriminatory, or abusive outputs

4.Human review

- editors should review generated story plots and character descriptions before use

5.Prompt constraints

- use instructions like “avoid stereotypes” or “write balanced and respectful characters”

5. Evaluation Methods

The system should be evaluated on both technical quality and creative usefulness.

Automatic evaluation

Some useful measures:

-perplexity for language fluency

- BLEU / ROUGE for comparison in limited structured tasks

- diversity metrics to avoid repeated plots

- coherence score for long text consistency

Human evaluation

This is the most important for creative writing.

Editors and writers can rate:

-creativity

- originality

- coherence

- character depth

- genre suitability

-usefulness in real writing work

Example evaluation questions

- Is the plot interesting?

- Are the characters believable?

- Is the content free from bias?

- Does the story maintain logical flow?

- Would a writer actually use this output?

6. Real-World Challenges

Building such a system in the real world is not easy. Some major challenges are:

a) Copyright and legal issues

Training on copyrighted books without permission can create legal problems.
So the publishing company must use:

- licensed datasets

- public-domain content

- internally approved training material

b) Lack of originality

The model may generate generic or repetitive plots.
This can reduce creative value.

c) Bias and stereotypes

The model may create unfair representations of people or communities if not carefully controlled.

d) Hallucination

The system may produce inconsistent facts or illogical events in the story.

e) Maintaining long context

In long stories, the model may forget earlier details such as:

- character personality

- timeline

- relationships

- setting rules

f) Overdependence on AI

Writers may rely too much on generated content, which can reduce human originality.

g) Genre control

The same model may struggle to perfectly match every genre unless it is fine-tuned well.

h) Editorial trust

Professional writers and editors may not trust AI outputs unless the results are consistently high quality.

4.The editor or writer reviews and edits the suggestion

5.Feedback is collected to improve the model over time
